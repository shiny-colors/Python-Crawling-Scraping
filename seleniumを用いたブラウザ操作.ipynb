```python
####seleniumを用いたブラウザ操作####
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import requests
from urllib.request import urlopen
import pandas as pd
import numpy as np
import time
import itertools
```

```python
#browserを読み込み検索窓にワードを入力して検索
def get_url(keyword):
    #seleniumでブラウザを読み込み
    browser = webdriver.Firefox()
    browser.get("https://www.google.com/")   #URLを検索
    
    #検索ワードを入力して検索
    browser.find_element_by_id("gsr").send_keys(keyword)   #検索ワード
    time.sleep(0.5)
    browser.find_element_by_id("gsr").send_keys(Keys.RETURN)
    #browser.find_element_by_name("btnK").click()   #こちらでもok検索する
    return browser
```

```python
#次ページに進む
def next_page(browser, wait_time1, wait_time2):
    time.sleep(wait_time1)
    try:
        flag = 0
        browser.find_element_by_css_selector("a#pnnext").click()
    except NoSuchElementException:
        flag = 1
    time.sleep(wait_time2)
    return flag
```

```python
#検索結果を取得
def get_search(browser):
    all_search = browser.find_elements_by_class_name("rc")
    n = len(all_search)
    title_list = [j for j in range(n)]
    url_list = [j for j in range(n)]
    display_url_list = [j for j in range(n)]
    description_list = [j for j in range(n)]

    for j in range(n):
        title_list[j] = all_search[j].find_element_by_tag_name("h3").text
        url_list[j] = all_search[j].find_element_by_tag_name("a").get_attribute("href")
        display_url_list[j] = all_search[j].find_element_by_tag_name("cite").text
        try:
            st = all_search[j].find_elements_by_class_name("st")
            if st!=[]:
                description_list[j] = st[0].text
            else:
                description_list[j] = print("")
        except NoSuchElementException:
            description_list[j] = ""
    
    #numpy arrayに変換
    title = np.array(title_list)
    url = np.array(url_list)
    display_url = np.array(display_url_list)
    description = np.array(description_list)
    return title, url, display_url, description   
```

```python
#任意のページまで検索結果を取得
pages = 10   #取得するページ数
browser = get_url("シンデレラガールズ")   #任意のキーワードで検索
time.sleep(0.5)   #待ち時間
title_list = [j for j in range(pages)]
url_list = [j for j in range(pages)]
display_url_list = [j for j in range(pages)]
description_list = [j for j in range(pages)]

for j in range(pages):
    #ページの情報を取得
    res = get_search(browser)
    title_list[j] = res[0]
    url_list[j] = res[1]
    display_url_list[j] = res[2]
    description_list[j] = res[3]

    #次のページへ
    flag = next_page(browser, 1.0, 0.5)
    if flag==1:
        break
```

```python
#リストを変換してdata frameに格納
title = np.array(list(itertools.chain(*[title_list[j] for j in range(pages)])))
url = np.array(list(itertools.chain(*[url_list[j] for j in range(pages)])))
display_url = np.array(list(itertools.chain(*[display_url_list[j] for j in range(pages)])))
description = np.array(list(itertools.chain(*[description_list[j] for j in range(pages)])))
pd.DataFrame({'1. タイトル' : title, '2. 概要' : description, '3. URL' : url, '4. Dispaly URL' : display_url})   #取得結果を表示
```
